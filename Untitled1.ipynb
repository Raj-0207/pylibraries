#Theoritical Questions

#Q1: What is NumPy, and why is it widely used in Python?
NumPy (Numerical Python) is a powerful library for numerical computing in Python. It provides support for multi-dimensional arrays, efficient array operations, and mathematical functions. NumPy is widely used because of its speed, efficiency, and ability to perform vectorized operations, which makes data processing much faster than using Python lists.

#Q2: How does broadcasting work in NumPy?
Broadcasting is a feature in NumPy that allows arrays of different shapes to be used in arithmetic operations without needing explicit replication. NumPy automatically expands the smaller array along the missing dimensions so that both arrays have compatible shapes, reducing memory usage and improving performance.

#Q3: What is a Pandas DataFrame?
A Pandas DataFrame is a two-dimensional labeled data structure similar to a table in a database or spreadsheet. It consists of rows and columns, where each column can hold different data types, making it useful for data manipulation, cleaning, and analysis.

#Q4: Explain the use of the groupby() method in Pandas.
The groupby() method in Pandas is used for grouping data based on one or more columns and applying aggregate functions like sum, mean, or count. It is useful for analyzing and summarizing large datasets.

#Q5: Why is Seaborn preferred for statistical visualizations?
Seaborn is preferred for statistical visualizations because it provides high-level API functions to create complex visualizations with minimal code. It integrates well with Pandas DataFrames and includes built-in themes and color palettes for visually appealing plots.

#Q6: What are the differences between NumPy arrays and Python lists?
NumPy arrays provide faster and more efficient data storage and computation than Python lists. They support element-wise operations, broadcasting, and have a fixed type for elements, whereas Python lists are flexible but slower due to their dynamic typing and lack of vectorized operations.

#Q7: What is a heatmap, and when should it be used?
A heatmap is a data visualization technique that represents data values as colors in a matrix format. It is useful for showing correlations between variables, detecting patterns, and visualizing large datasets effectively.

#Q8: What does the term “vectorized operation” mean in NumPy?
A vectorized operation refers to performing operations on entire arrays without using explicit loops. This enhances computational efficiency and speed, as NumPy optimizes these operations using low-level C implementations.

#Q9: How does Matplotlib differ from Plotly?
Matplotlib is a static plotting library that offers fine control over plot customization, while Plotly provides interactive visualizations with features like zooming, hovering, and dynamic updates, making it more suitable for web applications.

#Q10: What is the significance of hierarchical indexing in Pandas?
Hierarchical indexing allows multiple levels of indexes in a Pandas DataFrame, enabling better organization of data and facilitating complex queries, grouping, and reshaping operations.

#Q11: What is the role of Seaborn’s pairplot() function?
The pairplot() function in Seaborn creates scatter plots for all numerical variable combinations in a dataset, making it useful for exploring relationships between multiple features.

#Q12: What is the purpose of the describe() function in Pandas?
The describe() function provides summary statistics (mean, standard deviation, min, max, etc.) for numerical columns in a DataFrame, helping in understanding the data distribution.

#Q13: Why is handling missing data important in Pandas?
Handling missing data is crucial because missing values can distort analyses, cause errors, and impact model performance. Pandas provides functions like dropna() and fillna() to manage missing data effectively.

#Q14: What are the benefits of using Plotly for data visualization?
Plotly offers interactive, web-based visualizations with dynamic features like zooming and hovering, making it ideal for exploratory data analysis and presentations.

#Q15: How does NumPy handle multidimensional arrays?
NumPy provides the ndarray object, which supports multi-dimensional arrays with efficient storage, indexing, and mathematical operations.

#Q16: What is the role of Bokeh in data visualization?
Bokeh is a Python library for creating interactive and web-friendly visualizations, offering dynamic and scalable plots for large datasets.

#Q17: Explain the difference between apply() and map() in Pandas.
The apply() function applies a function to an entire DataFrame or Series, while map() is used specifically for transforming Series values.

#Q18: What are some advanced features of NumPy?
NumPy includes broadcasting, vectorization, linear algebra functions, Fourier transforms, and random number generation, enhancing numerical computation capabilities.

#Q19: How does Pandas simplify time series analysis?
Pandas provides functions for date handling, resampling, rolling statistics, and time-based indexing, making time series analysis straightforward.

#Q20: What is the role of a pivot table in Pandas?
A pivot table aggregates data based on multiple factors, making it useful for summarizing and analyzing large datasets.

#Q21: Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy array slicing is faster because arrays use contiguous memory storage and optimized operations, whereas lists require dynamic memory allocation.

#Q22: What are some common use cases for Seaborn?
Seaborn is commonly used for statistical data visualization, including correlation heatmaps, distribution plots, and categorical data analysis.

#1: How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(arr, axis=1)
print(row_sums)
```

#2: Write a Pandas script to find the mean of a specific column in a DataFrame.

```python
import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30, 40]})
mean_value = df['A'].mean()
print(mean_value)
```

#3: Create a scatter plot using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

#4: How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 3), columns=['A', 'B', 'C'])
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

#6: Create a DataFrame and add a new column based on an existing column.

#7: Write a program to perform element-wise multiplication of two NumPy arrays.

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print(result)
```

#8: Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

#9: Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30, 40]})
filtered_df = df[df['A'] > 20]
print(filtered_df)
```

#10: Create a histogram using Seaborn to visualize a distribution.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(100)
sns.histplot(data, bins=20, kde=True)
plt.show()
```

#11: Perform matrix multiplication using NumPy.
